```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic data (replace this with your actual data)
np.random.seed(42)
data = np.random.rand(100, 4)  # 100 samples with 4 features

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data)
print(clusters)
# Reduce dimensionality to 2D for plotting
pca_2d = PCA(n_components=2)
data_2d = pca_2d.fit_transform(data)

# Reduce dimensionality to 3D for plotting
pca_3d = PCA(n_components=3)
data_3d = pca_3d.fit_transform(data)

print(data_3d)

# 2D Scatter Plot
plt.figure(figsize=(8, 6))
for cluster in range(3):
    plt.scatter(data_2d[clusters == cluster, 0], data_2d[clusters == cluster, 1], label=f"Cluster {cluster}")
plt.title("2D PCA Clustering")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()

# 3D Scatter Plot
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111, projection='3d')
for cluster in range(3):
    ax.scatter(data_3d[clusters == cluster, 0], data_3d[clusters == cluster, 1], data_3d[clusters == cluster, 2], label=f"Cluster {cluster}")
ax.set_title("3D PCA Clustering")
ax.set_xlabel("PCA Component 1")
ax.set_ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
ax.legend()
plt.show()
```
