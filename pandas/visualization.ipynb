```python
# importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2D line plot
- Line Plot is used for Bivariate Analysis.
- Used in Categorical-numerical or numerical-numerical format
- Usecase: Time series data, trending data

```python
# plotting a simple function
price = [48000, 50000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2017, 2017, 2018, 2019, 2020, 2021]

plt.figure(figsize=(15,5)) # x: width, y: height
plt.plot(year, price)
plt.grid()
plt.title('Price per year')
plt.xlabel('Years') # categorical data always on the x-axis
plt.ylabel('Prices') # Numerical data on the y-asix
plt.show()
```

# From DataFrame

```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.figure(figsize=(10, 5))
plt.plot(batsman['index'], batsman['V Kohli'], label = 'Virat', color='green', linestyle = 'dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], label = 'Sharma', color='red', linestyle = 'dotted', marker='s', markersize=3)
plt.grid()
plt.title('Virat Kohli Vs RG Sharma Career Comparision')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend(loc='lower center')
plt.show()
```

```python
sns.lineplot(batsman, x='index', y='V Kohli', label='V Kohli')
sns.lineplot(batsman, x='index', y='RG Sharma', label='RG Sharma')
plt.xlabel('Years')
plt.ylabel('Runs/Scores')
plt.title("Score over time")
plt.grid()
plt.legend()
plt.show()
```

# Scatter Plots
- Used for Bivariate analysis
- Normally used for Numerical-Numerical and sometime Categorical-Numerical as well.
- Numerical-Numerical to find Correlation.

```python
x = np.linspace(-10, 10, 50)
y= 10*x+3+np.random.randint(0, 300, 50)
x
```

```python
y
```

```python
np.corrcoef(x, y)
```

```python
plt.scatter(x,y)
plt.show()
```

# Bar Charts
- Bivariate analysis
- Numerical vs categorical
- Usecase: Aggregate analysis of groups.

```python
plt.figure(figsize=(4, 6))
plt.bar(tips['day'], tips['total_bill'])
plt.show()
```

```python
tip = tips.groupby('day')[['total_bill']].mean()
```

```python
sns.barplot(tips, x='day', y='total_bill', hue='size', palette='deep')
```

```python
sns.barplot(tip, x=tip.index, y='total_bill')
```
