```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
expense = pd.read_csv('expense_data.csv')
expense
```

```python
expense.info()
```

```python
expense.dropna(axis = 1, inplace = True)
```

```python
expense.info()
```

```python
expense['Amount']== expense['Account.1'].any()
```

```python
expense['NRS']== expense['Account.1'].any()
```

```python
expense.drop(columns=['Amount', 'Currency', 'Account.1'], inplace = True)
```

```python
expense.info()
```

```python
expense['Date'] = pd.to_datetime(expense['Date'])
```

```python
expense
```

```python
expense['Income/Expense'].value_counts()
```

```python
expense['Year'] = expense['Date'].dt.year
expense['MOnth'] = expense['Date'].dt.month_name()
expense['Day'] = expense['Date'].dt.day
```

```python
exp = expense[expense['Income/Expense'] == 'Expense']
```

```python
exp
```

```python
exp_by_month = exp.groupby(['Year', 'MOnth'])['NRS'].sum()
```
