```python
provider_mean_purchase = ecom_purch.groupby('CC Provider')[['Purchase Price']].mean()
provider_mean_purchase
```

```python
plt.figure(figsize = (15,6))
ax = sns.barplot(ecom_purch, x = 'CC Provider', y = 'Purchase Price')
for container in ax.containers:
    ax.bar_label(container, fmt = '%.0f', label_type = 'edge', padding = 12)
plt.xticks(rotation = 70)
plt.show()
```

```python
provider_total_purchase = ecom_purch.groupby('CC Provider')[['Purchase Price']].sum()
provider_total_purchase
# provider_total_purchase = ecom_purch.groupby('CC Provider')[['Purchase Price']].sum().reset_index()
# provider_total_purchase
```

```python
plt.figure(figsize = (15,6))
ax = sns.barplot(provider_total_purchase, x = provider_total_purchase.index, y = 'Purchase Price')
for container in ax.containers:
    ax.bar_label(container, fmt = '%.0f', label_type = 'edge', padding = 12)
plt.xticks(rotation = 70)
plt.show()
```
